# one-hot编码练习   
## 一、 任务介绍   
* 1.给定一个词汇表(例如["cat", "dog", "bird", "fish"])   
* 2.编写函数转换词汇表对应的独热编码   
注意：处理边界--->不在词汇表的词返回提示和空值   
* 3. 测试： 为词汇表中的 "dog" 和不在词汇表中的 "tiger" 生成 One-Hot 向量   


## 二、 代码实现   
### 1. 为词汇表设置映射字典   
将每一个词对应一个索引：确保生成的编码唯一   

```python
import numpy as np
import pandas as pd
```

```python
# 映射函数
def build_vocab_map(vocab):
    # 去重
    vocab = set(vocab)
    # 排序使得固定顺序
    vocab = list(vocab)
    vocab.sort()

    word2int = {word : idx for idx,word in enumerate(vocab)}
    int2word = {idx : word for idx,word in enumerate(vocab)}
    return word2int,int2word
```

### 2.生成单个词的编码   
一个nx1维矩阵修改对应位置的值，注意检测边界   

```python
def build_one_hot(word,word2int):
    # 检查
    if word not in word2int:
        print(f"词{word}不在词汇表")
        return None
    
    #全0 n*1矩阵
    vocab_size = len(word2int)
    one_hot = np.zeros(vocab_size)
    one_hot[word2int[word]] = 1
    return one_hot
```

### 3.测试

```python

#定义词汇表
vocab = ["cat", "dog", "bird", "fish"]
word2int,int2word = build_vocab_map(vocab)

#输出编码
word = "dog"
vector = build_one_hot(word,word2int)

print(f"词\t{word}")
for w in vector.T:
    print(f"\t{int(w)}")

print("*"*50)
word1 = "tiger"
vec1 = build_one_hot(word1, word2int)

```
