# jieba中文使用
## 一、介绍   
### 1. 特点   
* 支持清晰分词模式：   
精确模式，试图将句子最精确地切开，适合文本分析；   
全模式，把句子中所有可以组成词的词汇都扫描出来，速度非常快，但不能解决歧义；   
搜索引擎模式，在精确模式的基础上，对长词再次切分，提高识别率，适合搜索引擎分词。   
* 支持繁体分词   
* 支持自定义词典   

### 2. 安装使用 
使用pip工具安装   
```bash
pip install jieba
```
使用   
```python
import jieba
```

### 算法介绍    
* 基于近邻实现的词图扫描，生成中汉句子所有可能成词情况所构成的有向无环图（DAG）   
* 采用动态规划寻找最大概率路径，根据词频的最大切分组合进行查找   
* 对于未登录词，采用了基于汉字成词能力的 HMM 模型，使用了 Viterbi 算法（最优路径算法）   

## 二、 功能实现
### 1. 导包   
导入包才可以使用   

````python
import jieba
````

### 2. 分词   
* *jieba.cut*--->接收3个参数:需要分词的字符串；*cut_all*参数用于控制是否采用全模式；HMM参数用于控制HMM模型   
* *jieba.cut_for_search*--->接收两个参数：需要分词的字符串；是否使用HMM模型。该方法适合用于引擎搜索构建倒排索引的分词（关键词->出现该词的文章列表），粒度比较细     
* 分词的字符串编码可以是 unicode 或 UTF-8 字符串、GBK 字符串。注意：不建议直接输入 GBK 字符串，可能无法未知地错误解码成 UTF-8   



````python
string = "我来到北京清华大学"
string1 = "小明硕士毕业于中国科学院计算所，后在日本京都大学深造"

#开启全模式
seg_list = jieba.cut(string,cut_all=True)
print("全模式: " + "/".join(seg_list))

# 精确模式   
seg_list = jieba.cut(string,cut_all=False)
print("精确模式: " + "/".join(seg_list))

#搜索引擎
seg_list = jieba.cut_for_search(string)
print("搜索引擎模式: " + "/".join(seg_list))


# 长文本
seg_list = jieba.cut(string1,cut_all=True)
print("全模式: " + "/".join(seg_list))

seg_list = jieba.cut(string1,cut_all=False)
print("精确模式: " + "/".join(seg_list))

seg_list = jieba.cut_for_search(string1)
print("搜索引擎模式: " + "/".join(seg_list))
````
