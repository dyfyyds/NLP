# jieba中文使用
## 一、介绍   
### 1. 特点   
* 支持清晰分词模式：   
精确模式，试图将句子最精确地切开，适合文本分析；   
全模式，把句子中所有可以组成词的词汇都扫描出来，速度非常快，但不能解决歧义；   
搜索引擎模式，在精确模式的基础上，对长词再次切分，提高识别率，适合搜索引擎分词。   
* 支持繁体分词   
* 支持自定义词典   

参考文档：[jieba教程](https://github.com/fxsjy/jieba)

### 2. 安装使用

```python
import jieba
```

### 算法介绍    
* 基于近邻实现的词图扫描，生成中汉句子所有可能成词情况所构成的有向无环图（DAG）   
* 采用动态规划寻找最大概率路径，根据词频的最大切分组合进行查找   
* 对于未登录词，采用了基于汉字成词能力的 HMM 模型，使用了 Viterbi 算法（最优路径算法）

### 2. 分词   
* *jieba.cut*--->接收3个参数:需要分词的字符串；*cut_all*参数用于控制是否采用全模式；HMM参数用于控制HMM模型   
* *jieba.cut_for_search*--->接收两个参数：需要分词的字符串；是否使用HMM模型。该方法适合用于引擎搜索构建倒排索引的分词（关键词->出现该词的文章列表），粒度比较细     
* 分词的字符串编码可以是 unicode 或 UTF-8 字符串、GBK 字符串。注意：不建议直接输入 GBK 字符串，可能无法未知地错误解码成 UTF-8   



````python
string = "我来到北京清华大学"
string1 = "小明硕士毕业于中国科学院计算所，后在日本京都大学深造"

#开启全模式
seg_list = jieba.cut(string,cut_all=True)
print("全模式: " + "/".join(seg_list))

# 精确模式   
seg_list = jieba.cut(string,cut_all=False)
print("精确模式: " + "/".join(seg_list))

#搜索引擎
seg_list = jieba.cut_for_search(string)
print("搜索引擎模式: " + "/".join(seg_list))


# 长文本
seg_list = jieba.cut(string1,cut_all=True)
print("全模式: " + "/".join(seg_list))

seg_list = jieba.cut(string1,cut_all=False)
print("精确模式: " + "/".join(seg_list))

seg_list = jieba.cut_for_search(string1)
print("搜索引擎模式: " + "/".join(seg_list))
````

### 3.自定义词典
开发者可以指定自己自定义的词典，以便包含 jieba 词库里没有的词。虽然 jieba 有新词识别能力，但是自行添加新词可以保证更高的正确率。   
* 单个词   
使用 add_word(word, freq=None, tag=None) 和 del_word(word) 可在程序中动态修改词典。   

使用 suggest_freq(segment, tune=True) 可调节单个词语的词频，使其能（或不能）被分出来。   
* 词典   
用法： jieba.load_userdict(file_name) # file_name 为文件类对象或自定义词典的路径   
词典格式和 dict.txt 一样，一个词占一行；每一行分三部分：词语、词频（可省略）、词性（可省略），用空格隔开，顺序不可颠倒。file_name 若为路径或二进制方式打开的文件，则文件必须为 UTF-8 编码。   
词频省略时使用自动计算的能保证分出该词的词频。   
```text
创新办 3 i
云计算 5
凱特琳 nz
台中
```

````python
string2 = "乒乓球拍卖完了吗"

#开启全模式
seg_list = jieba.cut(string2,cut_all=True)
print("全模式: " + "/".join(seg_list))

# 精确模式   
seg_list = jieba.cut(string2,cut_all=False)
print("精确模式: " + "/".join(seg_list))

#搜索引擎
seg_list = jieba.cut_for_search(string2)
print("搜索引擎模式: " + "/".join(seg_list))
````

精确模式分词将乒乓球分成一个词，可以使用词典导入的方式识别成'乒乓球拍'   

````python
# 将词频调大
jieba.add_word("乒乓球拍",freq=1000)   

seg_list = jieba.cut(string2,cut_all=False)
print("精确模式: " + "/".join(seg_list))

````
